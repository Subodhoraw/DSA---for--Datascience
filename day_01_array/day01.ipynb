```python
def remove_adjacent_duplicates(nums):
    if not nums:
        return 0
    
    # i is our writer the first hand 
    # it starts at index 0 
    i = 0
    # j is our explorer 
    # it starts with index 1 and start scanning  the whole list
    for j in range(1,len(nums)):
         #logic: if explorer find somthing diffrent writer will start work
         if nums[j] != nums[i]:
            i += 1 # writer will move to next slots 
            nums[i] = nums[j] # write the new unique value there 
    # return the clean portion of the list 
    #  i +1 will provide total count of unique elements found        
    return nums[:i +1]
# ---lets test it -----
raw_temps = [72,72,30,30,40,40,50,50,60,70,80]
cleaned_nums = remove_adjacent_duplicates(raw_temps)
print(f" original: {raw_temps}")
print(f"cleaned : {cleaned_nums}")
num = []
clean = remove_adjacent_duplicates(num)
print(f" while list is emplty :{clean}")




```
